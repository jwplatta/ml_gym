```python
def reverse_list(lst):
    left, right = 0, len(lst) - 1
    while left < right:
        lst[left], lst[right] = lst[right], lst[left]
        left += 1
        right -= 1
```

**Explanation:** We use two pointers starting at opposite ends. At each step, we swap the elements at `left` and `right`, then move both pointers inward. The loop terminates when the pointers meet or cross.

</details>
